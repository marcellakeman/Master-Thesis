```python
#Data2 = Data.head(1500)

def clean_text(text):
    #Clean text
    text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
    text = re.sub("[^a-zA-Z]", " ", text)# Search for all non-letters # Replace all non-letters with spaces
    text = text.lower() #Lower text
    text = re.sub('\[.*?\]', '', text)
    #text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text) #Remove numbers and text with numbers   
    text = ''.join(c for c in text if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    text = re.sub('\n', '', text) #Remove \n
    text = re.sub('€', '', text) #Remove €
    
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]

    #Remove stopwords
    #Dutch stop words
    stop = nltk.corpus.stopwords.words('dutch')
    text = [word for word in text if word not in stop]    
    
    # remove empty tokens and tokens with 1 
    text = [t for t in text if len(t) > 1]    
    
    return text

#Clean text data
#DataNL['Text Clean'] = DataNL['bericht tekst'].apply(lambda x: clean_text(x))

#Stem cleaned data
#dutchStemmer=SnowballStemmer("dutch")
#DataNL['Stemmed Clean'] = DataNL['Text Clean'].apply(lambda x: [dutchStemmer.stem(y) for y in x]) # Stem every word.

#Note: Stemming makes words incorrect

```

```python
def text_to_words(text):
    '''Transform a string to a list of words,
    removing all punctuation.'''
    text = text.lower()

    p = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    text = ''.join([ch for ch in text if ch not in p])

    return text.split()

def count_occurences(text, word_list):
    '''Count occurences of words from a list in a text string.'''
    text_list = text_to_words(text)

    intersection = [w for w in text_list if w in word_list]

    return len(intersection)

#Sustainable values
ENV_words = wordlist['Environmental Sustainability'].values
Social_words = wordlist['Social Sustainability'].values
Economic_words = wordlist['Economic Sustainability'].values

#Count the words per environmental/social/economic
DataNL['ENV_words'] = DataNL['Text Clean'].apply(count_occurences, args=(ENV_words, ))
DataNL['Social_words'] = DataNL['Text Clean'].apply(count_occurences, args=(Social_words, ))
DataNL['Economic_words'] = DataNL['Text Clean'].apply(count_occurences, args=(Economic_words, ))

#Label Sustainable Tweets

'''
DataNL['Sustainability'] = np.where((DataNL['ENV_words'] > ((DataNL['Social_words'])) & (DataNL['Economic_words'])) & 
                                    (DataNL['Sentiment'] >= 0), 'Environmental', 
                                     np.where((DataNL['Social_words'] > ((DataNL['ENV_words'])) & (DataNL['Economic_words'])) & 
                                              (DataNL['Sentiment'] >= 0), 'Social', 
                                               np.where((DataNL['Economic_words'] > ((DataNL['ENV_words'])) & (DataNL['Social_words'])) & 
                                                        (DataNL['Sentiment'] >= 0), 'Economic', 'Non-Sustainable')))

'''

DataNL['Sustainability'] = np.where((DataNL['ENV_words'] > ((DataNL['Social_words'])) & (DataNL['Economic_words'])) & 
                                     (DataNL['Sentiment'] >= 0), 'Positive Environmental', 
                                np.where((DataNL['ENV_words'] > ((DataNL['Social_words'])) & (DataNL['Economic_words'])) & 
                                         (DataNL['Sentiment'] < 0), 'Negative Environmental', 
                                    np.where((DataNL['Social_words'] > ((DataNL['ENV_words'])) & (DataNL['Economic_words'])) & 
                                             (DataNL['Sentiment'] >= 0), 'Positive Social', 
                                        np.where((DataNL['Social_words'] > ((DataNL['ENV_words'])) & (DataNL['Economic_words'])) & 
                                                 (DataNL['Sentiment'] < 0), 'Negative Social', 
                                            np.where((DataNL['Economic_words'] > ((DataNL['ENV_words'])) & (DataNL['Social_words'])) & 
                                                     (DataNL['Sentiment'] >= 0), 'Positive Economic', 
                                                np.where((DataNL['Economic_words'] > ((DataNL['ENV_words'])) & (DataNL['Social_words'])) & 
                                                         (DataNL['Sentiment'] < 0), 'Negative Economic', 
                                                    'Non-Sustainable'                                 
                                    ))))))


#New dummy columns to groupby later on

DataNL.loc[DataNL['Sustainability'] == 'Positive Environmental', 'Positive Environmental'] = 1
DataNL.loc[DataNL['Sustainability'] == 'Negative Environmental', 'Negative Environmental'] = 1

DataNL.loc[DataNL['Sustainability'] == 'Positive Social', 'Positive Social'] = 1
DataNL.loc[DataNL['Sustainability'] == 'Negative Social', 'Negative Social'] = 1

DataNL.loc[DataNL['Sustainability'] == 'Positive Economic', 'Positive Economic'] = 1
DataNL.loc[DataNL['Sustainability'] == 'Negative Economic', 'Negative Economic'] = 1


DataNL
```

```python
#Get number of words

from sklearn.feature_extraction.text import CountVectorizer

def clean_text(text):
    #Clean text
    text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
    text = re.sub("[^a-zA-Z]", " ", text)# Search for all non-letters # Replace all non-letters with spaces
    text = text.lower() #Lower text
    text = re.sub('\[.*?\]', '', text)
    #text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text) #Remove numbers and text with numbers   
    text = ''.join(c for c in text if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    text = re.sub('\n', '', text) #Remove \n
    text = re.sub('€', '', text) #Remove €
    
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]

    #Remove stopwords
    #Dutch stop words
    stop = nltk.corpus.stopwords.words('dutch')
    text = [word for word in text if word not in stop]    
    
    # remove empty tokens and tokens with 1 
    text = [t for t in text if len(t) > 1]    
    
    return text

#Vectorize words
countVectorizer = CountVectorizer(analyzer=clean_text) #No word stemming in clean_text
countVector = countVectorizer.fit_transform(DataNL['bericht tekst'])
print('{} Number of tweets has {} words'.format(countVector.shape[0], countVector.shape[1]))

#Dataframe of words
count_vect_df = pd.DataFrame(countVector.toarray(), columns=countVectorizer.get_feature_names())
count_vect_df.head()
```
